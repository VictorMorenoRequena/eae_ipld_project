```python
from PIL import Image

# Example crop dimensions, modify them to test your code
crop_min_h = 600
crop_max_h = 1000
crop_min_w = 550
crop_max_w = 1100

# Get the image size
image_path = '/Users/victormorenorequena/Downloads/EAE/Programming/eae_ipld_project/data/starry_night.png'  # Replace with the path to your image
img = Image.open(image_path)
img_size = img.size

# Constraint checks
if not (0 < crop_min_h < crop_max_h < img_size[1]):
    print("Error: Invalid crop height dimensions.")

if not (0 < crop_min_w < crop_max_w < img_size[0]):
    print("Error: Invalid crop width dimensions.")
```

```python
#1.3, display the cropped image
# Generate the crop array using NumPy array slicing
cropped_img_arr = img_arr[crop_min_h:crop_max_h, crop_min_w:crop_max_w, :]

# Display the cropped image (you can use Matplotlib for this)
plt.imshow(cropped_img_arr)
plt.title('Cropped Image')
plt.show()
```

```python
# Ex 1.4: Saving the cropped image into a new file, set the filename of the new image

cropped_img = Image.fromarray(cropped_img_arr)

# Set the filename of the new image
output_file_name = '/Users/victormorenorequena/Downloads/EAE/Programming/eae_ipld_project/data/cropped_starry_night.png'  # Replace with your desired output file name

# Save the cropped image as a new file
cropped_img.save(output_file_name) 

print(f"Cropped image saved as {output_file_name}")
```
